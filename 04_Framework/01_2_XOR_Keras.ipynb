## XOR 이해하기 with Keras

```python
# tensorflow와 tf.keras를 임포트합니다
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
from tensorflow import keras
```

```python
tf.__version__
```

```python
# 헬퍼(helper) 라이브러리를 임포트합니다
import numpy as np
import matplotlib.pyplot as plt
```

Remember !! XOR : 모두 같으면 0

```python
x_data = [[0, 0],
          [0, 1],
          [1, 0],
          [1, 1]]

y_data = [0,
          1,
          1,
          0]

x_data = np.array(x_data, dtype=np.float32)
y_data = np.array(y_data, dtype=np.float32)
```

```python
plt.scatter(x_data[:, 0], x_data[:, 1], c=y_data)
```

```python
from tensorflow.keras import layers
from tensorflow.keras import activations
from tensorflow.keras import optimizers
from tensorflow.keras import models
```

```python
model = models.Sequential()
model.add(layers.Dense(2, input_dim=2)) #인풋레이어 : 입력 노드값이 2 출력 노드값이 2
model.add(layers.Activation('tanh'))
#model.add(layers.Dense(2)) # 출력 노드값 그대로 입력받음(2) 출력 노드값 2
#model.add(layers.Activation('tanh')) # 활성화 함수로 tanh 사용
model.add(layers.Dense(1)) #출력 노드값 그대로 입력받음(2) 출력 노드값 1
model.add(layers.Activation('sigmoid')) #활성화 함수로 sigmoid 사용
```

```python
sgd = optimizers.SGD(lr=0.1)
model.compile(loss='binary_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```python
model.summary()
```

```python
from tensorflow.keras.utils import plot_model
plot_model(model, to_file='model_xor.png', show_shapes=True)
```

```python
history = model.fit(x_data, y_data,  batch_size=1, epochs=500)
```

### 학습 결과 그려보기


```python
plt.plot(history.history['loss'])
plt.title('Loss')
plt.xlabel('epoch')
```

```python
plt.plot(history.history['acc'])
plt.title('Training accuracy')
plt.xlabel('epoch')
```

### 확률 찍어보기

```python
hypothesis = model.predict(x_data)
print(hypothesis)
```

```python
predicted = hypothesis > 0.5
print(predicted)
```
